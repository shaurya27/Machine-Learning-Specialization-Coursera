```python
def predict_output(feature_matrix, weights):
    output=np.dot(feature_matrix,weights)
    return(output)
```

```python
def feature_derivative(errors, feature,weights,l2_penalty):
    derivative = 2*np.dot(errors, feature) + 2*l2_penalty*weights
    return(derivative)
```

```python
def regression_gradient_descent(feature_matrix,output,initial_weights,step_size,l2_penalty,max_iterations=100):
    weights = initial_weights
    prediction=predict_output(feature_matrix,weights)
    gradient_magnitude=0
    iteration=0
    converged=False
    while not converged:
        prediction=predict_output(feature_matrix,weights)
        error=prediction-output
        gradient_sum_square=0
        for i in range(len(weights)):
            if i==0:
                derivative=2*np.dot(error, feature_matrix[:,i])
            else:    
                derivative=feature_derivative(error, feature_matrix[:,i],weights[i],l2_penalty)
            gradient_sum_square+= derivative*derivative
            weights[i]-=(step_size*derivative)
        gradient_magnitude=sqrt(gradient_sum_square)
        iteration+=1
        print weights
        print gradient_magnitude
        if (iteration > max_iterations):
            converged=True
    return(weights)
       
```

```python
simple_weights_0_penalty=regression_gradient_descent(feature_sqft,train_data['price'],np.array([0.0,0.0]),1e-12,0.0,max_iterations=1000)
```

```python
simple_weights_penalty=regression_gradient_descent(feature_sqft,train_data['price'],simple_weights_0_penalty,1e-12,1e11,max_iterations=1000)
```
